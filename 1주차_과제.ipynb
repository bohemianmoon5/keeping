##<음성>
* 타입캐스트(typecast)
 * 인공지능 음성 생성 기술을 기반으로, 50여 개의 다양한 음성을 실시간 오디오 컨텐츠로 변환해 주는 서비스 
 * 전문 성우의 음성+TTS(Text to Speech) 
* 클로바 해피콜(CLOVA HappyCall)
 * 완전판매 여부 확인을 위한 사후 모니터링을 비롯해, 아웃바운드 전화 업무를 수행하는 솔루션
 * 자연어처리, 음성 인식 및 합성, 텍스트 분석 등의 기술 사용
* 구글 홈(Google Home)
 * AI 음성비서 기술인 구글 어시스턴트를 기반으로 하는 AI 스피커
 * 다중언어, 보이스 매치, 멀티룸 모드 등 다양한 기능 제공
* 샬롯홈
 * 상품들을 직접 보고 손가락으로 터치하면서 목소리로 주문하는 서비스
 * 날씨, 유튜브, 팟캐스트 등 총 18가지 컨텐츠 이용 가능

##<이미지>
* 얼굴인식 출입 서비스(LG CNS)
 * 인공지능(AI) 기술로 얼굴 정보를 분석, 신분 인증 및 출입을 제어하는 서비스
* aiBee BUILDER
 * 이미지 객체 검출, 이미지 분류, 이미지 유사검색 등의 학습을 지원
* 페이스미(FaceMe)
 * 표정을 인식해 사람의 나이와 성별, 감정을 식별
* 신한 Face Pay
 * 얼굴 인식만으로 결제를 하는 방식의 결제 시스템

##<자율주행>
* 오토파일럿(Autopilot)
 * 테슬라의 전기차 '모델S'에 장착된 자율주행 소프트웨어
 * 아직은 부분적 자율주행만 가능
* 슈퍼 크루즈(Super Cruise)
 * 자율주행 시스템에 스스로 차선을 변경할 수 있는 기능을 더함
 * 오는 하반기에 적용 예정
* XP7102
 * 직진과 회전이 가능한 자율주행 트랙터(2022년 상용화 예정)
 * 설정한 작업을 자동으로 할 수 있고, 실시간으로 작업위치 확인이 가능
* 에어스타
 * 인공지능 챗봇서비스, 자율주행·음성인식 등 첨단 ICT 기술을 접목한 안내로봇
 * 인천국제공항에서 운영 중
